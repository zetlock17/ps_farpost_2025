```python
import sqlite3
import pandas as pd
DB_PATH = '../databases/dataset.db'
conn = sqlite3.connect('../databases/dataset.db')

query = """
SELECT name FROM sqlite_master 
WHERE type='table' 
ORDER BY name
"""
tables = pd.read_sql_query(query, conn)['name'].tolist()

print(f"НАЙДЕНО {len(tables)} ТАБЛИЦ:")
for i, table in enumerate(tables, 1):
    print(f"   {i}. {table}")
print()

print("первые 3 строки каждой таблицы")

for table_name in tables:
    try:
        # Количество строк в таблице
        count_df = pd.read_sql_query(f'SELECT COUNT(*) as total FROM {table_name}', conn)
        total_rows = count_df.iloc[0]['total']
        
        # Первые 3 строки
        df = pd.read_sql_query(f'SELECT * FROM {table_name} LIMIT 3', conn)
        
        print(f"\n{'='*60}")
        print(f"ТАБЛИЦА: {table_name.upper()}")
        print(f"Всего строк: {total_rows:,}")
       

        if len(df) > 0:
            print(df.to_string(
                index=False, 
                max_colwidth=100,
                float_format='%.2f'
            ))
        else:
            print("ПУСТАЯ ТАБЛИЦА")
            
        
    except Exception as e:
        print(f"ОШИБКА в {table_name}: {e}\n")


conn.close()
```

```python
import sqlite3
DB_PATH = '../databases/dataset.db'
def count_null_folk_district_real_buildings(db_path):
    try:
        # Подключаемся к базе данных
        conn = sqlite3.connect(db_path)
        cursor = conn.cursor()
        
        # Выполняем запрос
        cursor.execute("""
            SELECT COUNT(*) 
            FROM buildings 
            WHERE folk_district_id IS NULL AND is_fake = 0
        """)
        
        # Получаем результат
        count = cursor.fetchone()[0]
        
        print(f"Количество НЕ фейковых зданий (is_fake=0) с folk_district_id == NULL: {count}")
        
        # Закрываем соединение
        conn.close()
        
        return count
        
    except sqlite3.Error as e:
        print(f"Ошибка при работе с базой данных: {e}")
        return None

count_null_folk_district_real_buildings(DB_PATH)
```

```python
import sqlite3
DB_PATH = '../databases/dataset.db'

def show_folk_districts(db_path):
    try:
        conn = sqlite3.connect(db_path)
        cursor = conn.cursor()
        
        # Получаем общее количество
        cursor.execute("SELECT COUNT(*) FROM folk_districts")
        count = cursor.fetchone()[0]
        print(f"Всего записей в folk_districts: {count}\n")
        
        # Получаем все записи
        cursor.execute("SELECT id, name FROM folk_districts")
        districts = cursor.fetchall()
        
        print("Список народных районов:")
        print("-" * 50)
        for i, (district_id, district_name) in enumerate(districts, 1):
            print(f"{i:3d}. ID: {district_id} | Название: {district_name}")
        
        conn.close()
        
        return districts
        
    except sqlite3.Error as e:
        print(f"Ошибка при работе с базой данных: {e}")
        return None

show_folk_districts(DB_PATH)
```
